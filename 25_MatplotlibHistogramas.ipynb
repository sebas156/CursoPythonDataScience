```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parametros de la distribucion normal (media, desviacion estandar, cantidad de datos)
data = np.random.normal(170, 10, 200)
print(data)
```

```python
plt.hist(data, bins=20, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Distribucion de alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
np.random.seed(0)

ages = [np.random.normal(24, 5, 100),
        np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100)]

print(ages)
```

```python
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Distribucion de edades')
plt.xlabel('Grupo')
plt.ylabel('Edad(Años)')
plt.show()
```

## Con patch_artist=False

```python
plt.boxplot(ages, patch_artist=False, notch=True, vert=True, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Distribucion de edades')
plt.xlabel('Grupo')
plt.ylabel('Edad(Años)')
plt.show()
```

## notch=False

```python
plt.boxplot(ages, patch_artist=True, notch=False, vert=True, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Distribucion de edades')
plt.xlabel('Grupo')
plt.ylabel('Edad(Años)')
plt.show()
```
