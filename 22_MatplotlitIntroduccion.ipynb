```python
import numpy as np
import matplotlib.pyplot as plt
```

## Diagrama de lineas

```python
month = np.array(['E','F','M','A','May'])
sales = np.array([20,25,30,28,35])
```

```python
#Configurar el tamaño de la figura
plt.figure(figsize=(8,4))

#Crear el gráfico
plt.plot(month,sales,marker='o',color='blue',linewidth=2,markersize=10)
plt.title('Ventas mensuales')
plt.xlabel('Meses')
plt.ylabel('Ventas')
```

## Diagrama de dispersion

```python
plt.figure(figsize=(8,6))

hours = [2,3,4,5,6,7,8]
exam = [55,60,65,70,75,80,85]

plt.scatter(hours,exam,marker='o',color='red',s=100)
plt.title('Relación entre horas de estudio y calificación')
plt.xlabel('Horas de estudio')
plt.ylabel('Puntaje')

plt.show()
```
