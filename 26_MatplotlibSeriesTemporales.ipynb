```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.dates import DateFormatter
```

```python
dates = pd.date_range(start='2023-01-01', periods=100)
values = np.random.rand(100).cumsum()
data = pd.DataFrame({'date': dates, 'value': values})
```

## Grafico de lineas

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(data['date'], data['value'],color='green')

plt.xticks(rotation=45)

ax.set_title('Serie de tiempo con formato en las fechas')
ax.set_xlabel('Fechas')
ax.set_ylabel('Valores')

plt.show()
```

```python
data['Year'] = data['date'].dt.year
data['Month'] = data['date'].dt.month
data['Day'] = data['date'].dt.day

sales_grouped_by_month = data.groupby(by='Month', as_index=False).agg({'value':'sum'})
print(sales_grouped_by_month)
```

```python
plt.plot(sales_grouped_by_month['Month'],sales_grouped_by_month['value'],color='blue', marker = 'o')
plt.title('Total sales by Month')
plt.xlabel('Month')
plt.ylabel('Total sales')

plt.show()
```

```python
dates = pd.date_range(start='2023-01-01', periods=12, freq='M')
sales = np.random.randint(1000,5000, size=12)
```

```python
sales_data = pd.DataFrame({'dates':dates,'sales':sales})

plt.plot(sales_data['dates'],sales_data['sales'],marker='o',linestyle='-', label='Ventas mensuales')


plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))

plt.xticks(rotation=45)

plt.title('Analisis de ventas mensuales')
plt.xlabel('Fechas')
plt.ylabel('Ventas')
plt.legend()

plt.tight_layout()

plt.show()
```
