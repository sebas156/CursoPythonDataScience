```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import pandas as pd
```

### 1. **Creación de Subplots Básicos**

El método plt.subplot() o plt.subplots() es la base para crear múltiples gráficos en una sola figura.

#### Ejemplo básico:

```python
# Datos de ejemplo

x = np.linspace(0, 10, 100)

y1 = np.sin(x)

y2 = np.cos(x)



# Crear una figura con dos subplots (1 fila, 2 columnas)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))



# Gráfico 1: Seno

ax1.plot(x, y1, 'r')

ax1.set_title("Seno")



# Gráfico 2: Coseno

ax2.plot(x, y2, 'b')

ax2.set_title("Coseno")



plt.show()
```

### 2. **Personalización de los Subplots**

#### Títulos, etiquetas y leyendas:

Cada subplot es un eje individual que se puede personalizar de manera independiente.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))



# Gráfico 1: Seno

ax1.plot(x, y1, 'r')

ax1.set_title("Seno")

ax1.set_xlabel("Tiempo")

ax1.set_ylabel("Amplitud")



# Gráfico 2: Coseno

ax2.plot(x, y2, 'b')

ax2.set_title("Coseno")

ax2.set_xlabel("Tiempo")

ax2.set_ylabel("Amplitud")



plt.show()
```

### 3. **Ajuste del Espaciado entre Subplots**

Cuando los gráficos están muy juntos, puedes ajustar el espaciado entre ellos con plt.subplots\_adjust().

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))



# Agregar los gráficos

ax1.plot(x, y1, 'r')

ax1.set_title("Seno")

ax1.set_xlabel("Tiempo")

ax1.set_ylabel("Amplitud")



ax2.plot(x, y2, 'b')

ax2.set_title("Coseno")

ax2.set_xlabel("Tiempo")

ax2.set_ylabel("Amplitud")



# Ajustar el espaciado

plt.subplots_adjust(wspace=0.3)  # Aumentar el espacio horizontal entre subplots



plt.show()
```

### 4. **Crear Subplots con Diferentes Tamaños (GridSpec)**

GridSpec permite crear layouts complejos donde cada subplot puede ocupar un número variable de filas o columnas.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.randn(100)

gs = gridspec.GridSpec(2, 2, width_ratios=[1, 1], height_ratios=[2, 1])
fig = plt.figure(figsize=(10, 8))

# Primer subplot Grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y,color='red')
ax1.set_title('Seno de x')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

# Primer subplot Grande, ocupa la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(data,bins=20,color='purple',edgecolor='black') 
ax2.set_title('Histograma de datos')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')

# Primer subplot Grande, ocupa la esquina inferior derecha
ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y,color='red')
ax3.set_title('Dispersión de datos')
ax3.set_xlabel('X')
ax3.set_ylabel('sin(x)')

plt.tight_layout()
plt.show()
```
