```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
sales = [300, 500, 700, 200]

#Creacion del grafico de barras verticales
plt.bar(categories, sales, color='skyblue',label='Ventas mensuales')

#Anotacion utilizando una flecha par destacar un punto especifico
plt.annotate('Pico de ventas', xy=('Producto C', 700),xytext=('Producto D',710), arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Ventas mensuales de productos')
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
sales = [300, 500, 700, 200]

#Creacion del grafico de barras verticales
plt.barh(categories, sales, color='skyblue',label='Ventas mensuales')

plt.title('Ventas mensuales de productos')
plt.ylabel('Productos')
plt.xlabel('Ventas')
plt.legend()

plt.show()
```

## Diagrama de Pastel

```python
plt.figure(figsize=(8, 6))

categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
sales = [300, 500, 700, 200]

#Creacion del diagrama de pastel
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'yellow', 'green', 'red'])

#Titulo
plt.title('Ventas mensuales de productos')

#Asegurar que el grafico sea un circulo perfecto
plt.axis('equal')

#Mostar el grafico
plt.show()
```
