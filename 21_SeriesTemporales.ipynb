```python
import pandas as pd
```

```python
path = r'C:\Users\seboc\OneDrive\Escritorio\PROJECTS\DATASCIENCE\CursoPythonDataScience\Online_Retail.csv'
df_retail_data = pd.read_csv(path,encoding='windows-1252')
df_retail_data.head()
```

```python
df_retail_data.info()
```

```python
df_retail_data['InvoiceDate'] = pd.to_datetime(df_retail_data['InvoiceDate'])
df_retail_data.info()
```

```python
df_retail_data.dropna(subset=['InvoiceDate'],inplace=True)
```

```python
df_retail_data.set_index('InvoiceDate')
```

```python
df_retail_data.set_index('InvoiceDate',inplace=True)
```

```python
df_retail_data['Year'] = df_retail_data.index.year
df_retail_data['Month'] = df_retail_data.index.month
df_retail_data['Day'] = df_retail_data.index.day
df_retail_data['Weekday'] = df_retail_data.index.weekday
df_retail_data['Hour'] = df_retail_data.index.hour

print(df_retail_data['Year'].head())
print(df_retail_data['Month'].head())
print(df_retail_data['Day'].head())
```

```python
df_retail_data
```

## Utilizando loc

```python
df_retail_data_2011 = df_retail_data.loc['2011']
print(df_retail_data_2011)
```

## Crear rangos de fechas

```python
date_range = pd.date_range(start='2011-12-01',end='2011-12-15',freq='D')
print(date_range)
```

```python
df_dates = pd.DataFrame(date_range,columns=['Date'])
print(df_dates)
```
